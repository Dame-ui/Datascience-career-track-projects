**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great!

#### The below cell is used to define a ***helper function*** for calculating the ratio of average home prices in (2018/1998)

```python
dfg[dfg['London_borough'] == 'Hackney']
```

```python
def create_price_ratio(dfg_filter_borough):
    
    # definition of price_1998 and price_2018
    price_1998 = float(dfg_filter_borough[dfg_filter_borough['Date_Year'] == 1998]['Avg_home_price'])
    price_2018 = float(dfg_filter_borough[dfg_filter_borough['Date_Year'] == 2018]['Avg_home_price'])
    
    ratio_prices = price_2018 / price_1998
    
    return (ratio_prices)
    
# An example fxn call

#printing with .format()
"The ratio of avg_home_price (2018/1998) for the greenwich borrough is: {}".format(create_price_ratio(dfg[dfg["London_borough"] == 'Greenwich'])) 

#printing with f"str{}
f"The ratio of avg_home_price (2018/1998) for the greenwich borrough is: {create_price_ratio(dfg[dfg['London_borough'] == 'Greenwich'])}" 
```

#### The below cell(s) are used to iterate and traverse through the unique london boroughs column and calculate the ratio of average home prices in (1998/2018) for each borough. The unique borrow names are the keys of the ***ratios dict*** and the corresponding ratio values are the values of the dict.

```python
ratios = {}
for b in dfg['London_borough'].unique():
    
    df_new = dfg[dfg['London_borough'] == b]
    ratios[b] = create_price_ratio(df_new)
    
print(ratios)
```

### The below cell(s) are used to convert the list of the single ratios dict into a useable dataframe and to transpose the dataframe

```python
ratios_dfg = pd.DataFrame([ratios])
ratios_dfg
```
